```python
import pandas as pd

df = pd.read_csv('Sample - Superstore.csv', encoding='latin1')
df.head()
```

```python
df.info()
df.describe()
df.columns
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce')
df['Ship Date'] = pd.to_datetime(df['Ship Date'], errors='coerce')
```

```python
df['Month'] = df['Order Date'].dt.to_period('M')
```

```python
monthly_sales = df.groupby('Month')['Sales'].sum()
monthly_sales.plot(kind='line', title='Monthly Sales Trend', figsize=(10, 5))
```

```python
top_subcats = df.groupby('Sub-Category')['Sales'].sum().sort_values(ascending=False).head(10)
top_subcats.plot(kind='bar', title='Top 10 Sub-Categories by Sales', figsize=(10, 5))
```

```python
region_profit = df.groupby('Region')['Profit'].sum().sort_values()
region_profit.plot(kind='barh', title='Profit by Region', figsize=(8, 4), color='green')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='Discount', y='Profit', hue='Category')
plt.title('Discount vs Profit')
plt.grid(True)
```

```python
import plotly.express as px

# Discount vs Profit Scatter
fig1 = px.scatter(df, x='Discount', y='Profit', color='Category',
                  title='Discount vs Profit by Category', hover_data=['Product Name'])
fig1.show()

# Monthly Sales Trend
df['Month'] = df['Order Date'].dt.to_period('M').astype(str)
monthly_sales = df.groupby('Month')['Sales'].sum().reset_index()

fig2 = px.line(monthly_sales, x='Month', y='Sales', title='Monthly Sales Trend')
fig2.show()
v
```
